# E-commerce Returns & Delivery Analysis (1M+ Records)

**Dataset:** 1,000,000 e-commerce transaction records across 20 fields (category, brand, price, seller info, delivery status, returns, etc.)

**Goal:** Identify which factors, if any, are associated with order returns and delivery delays.

**Tools:** Python (pandas) for cleaning and analysis.

## Step 1: Data Profiling

Checked shape, data types, nulls, and summary statistics before any cleaning.

## Step 2: Data Quality Findings

- 1,000,000 rows, 20 columns, **zero missing values** across all fields
- `purchase_date` was stored as text; converted to datetime
- Numeric fields (price, discount, final_price, rating, seller_rating) all fall within logical ranges — no negative prices, no out-of-range ratings
- `final_price` was verified against `price` and `discount`; differences were negligible (rounding only)
- **Brand and category fields are independently randomized** — every one of the 12 brands appears across all 5 categories, confirming synthetic data generation rather than real product-brand relationships. Brand-category pairs are therefore not treated as meaningful signal in this analysis.
- Categorical fields (location, device, payment_method, delivery_status) are consistently formatted with no typos or casing issues

**Conclusion:** minimal cleaning was required beyond the datetime conversion.

## Step 3: Business Questions

1. Which category has the highest return rate?
2. Which location has the highest delivery delay rate?
3. Does seller rating correlate with returns?
4. Does shipping time correlate with returns?
5. Does payment method correlate with returns?

**Finding:** Return rates are nearly identical across categories (11.3%–11.8%). No category shows a meaningfully elevated return risk.

**Finding:** Delay rates are nearly identical across all 5 cities (29.2%–29.8%). Location has no meaningful effect on delivery delays.

**Finding:** Average seller rating is virtually identical for returned vs. non-returned orders (3.75 vs 3.75). Seller rating does not predict returns.

**Finding:** Orders that were returned had a noticeably longer average shipping time (3.41 days) compared to orders that were not returned (3.13 days) — roughly a 9% relative difference. This is the strongest signal found in the analysis.

**Finding:** Return rates are nearly identical across all payment methods (11.5%–11.7%). Payment method has no meaningful effect on returns.

## Final Conclusion

Across five variables tested — category, location, seller rating, payment method, and shipping time — only **shipping time** showed a meaningful relationship with returns. Orders that were eventually returned had ~9% longer average shipping times than orders that were not returned. All other variables tested showed flat, statistically negligible differences.

**Business takeaway:** Returns in this dataset appear to be driven primarily by delivery experience (shipping delays) rather than product category, seller quality, customer location, or payment method. This suggests operational/logistics improvements (faster, more reliable shipping) may have more impact on reducing returns than seller vetting or category-specific interventions.
